```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import altair as alt

filename = 'ds4200_data.csv'
df = pd.read_csv(filename)
df.columns.tolist()
```

```python
df = df.rename(columns={
    'How old are you?': 'Age',
    'How would you best describe yourself?': 'Sex',
    'Are you currently a student?': 'Student',
    'What city do you live in?': 'City',
    'For the following list of movies, please select if/where you first watched each movie. [Wicked]': 'Wicked',
    'For the following list of movies, please select if/where you first watched each movie. [Zootopia 2]': 'Zootopia 2',
    'For the following list of movies, please select if/where you first watched each movie. [Sinners]': 'Sinners',
    'For the following list of movies, please select if/where you first watched each movie. [Oppenheimer]': 'Oppenheimer',
    'For the following list of movies, please select if/where you first watched each movie. [Superman]': 'Superman',
    'For the following list of movies, please select if/where you first watched each movie. [No Other Choice]': 'No Other Choice',
    'For the following list of movies, please select if/where you first watched each movie. [One Battle After Another]': 'One Battle After Another',
    'For the following list of movies, please select if/where you first watched each movie. [Dune 2]': 'Dune 2',
    'For the following list of movies, please select if/where you first watched each movie. [Marty Supreme]': 'Marty Supreme',
    'For the following list of movies, please select if/where you first watched each movie. [Barbie]': 'Barbie',
    'How many movies have you seen in theaters in the past 3 months (estimate)?': '# of Movies Seen in Theaters in Past 3 Months',
    'How many movies have you watched on streaming in the past 3 months (estimate)?': '# of Movies Watched on Streaming in Past 3 Months',
    'Which streaming services are you subscribed to? (Select all that apply)': 'Streaming Services',
    'What is your biggest barrier to going to the movie theater?': 'Biggest Barrier to Going to Movie Theater'
})
df.head()
```

```python
movie_cols = [
    'Wicked', 'Zootopia 2', 'Sinners', 'Oppenheimer',
    'Superman', 'No Other Choice', 'One Battle After Another',
    'Dune 2', 'Marty Supreme', 'Barbie'
]

df_long = df.melt(
    id_vars=['Age', 'Sex', 'Student', 'City'],
    value_vars=movie_cols,
    var_name='Movie',
    value_name='Response'
)

df_long = df_long[df_long['Response'] != 'Did not watch']
```

```python
movie_info = pd.DataFrame({
    'Movie': [
        'Wicked', 'Zootopia 2', 'Sinners', 'Oppenheimer',
        'Superman', 'No Other Choice', 'One Battle After Another',
        'Dune 2', 'Marty Supreme', 'Barbie'
    ],
    'Genre': [
        'Musical', 'Animation', 'Thriller', 'Drama',
        'Superhero', 'Drama', 'Drama',
        'Sci-Fi', 'Drama', 'Comedy'
    ],
    'Year': [
        2024, 2025, 2024, 2023,
        2025, 2024, 2024,
        2024, 2024, 2023
    ]
})

df_long = df_long.merge(movie_info, on='Movie')
```

```python
mode_summary = (
    df_long
    .groupby(['Genre', 'Response'])
    .size()
    .reset_index(name='Count')
)

import altair as alt

alt.Chart(mode_summary).mark_bar().encode(
    x=alt.X('Genre:N', title='Genre'),
    y=alt.Y('Count:Q', title='Number of Responses'),
    color=alt.Color('Response:N', title='Viewing Type'),
    tooltip=['Genre', 'Response', 'Count']
).properties(
    title='Theater vs Streaming by Genre'
)
```

```python
df_long['AgeGroup'] = pd.cut(
    df_long['Age'],
    bins=[0,18,25,35,50,65,100],
    labels=['<18','18-25','26-35','36-50','51-65','65+']
)

df_theater = df_long[df_long['Response'] == 'Watched in theaters']

sex_summary = (
    df_theater.groupby('Sex')
    .size()
    .reset_index(name='Count')
)

age_summary = (
    df_theater.groupby('AgeGroup')
    .size()
    .reset_index(name='Count')
)
```

```python
alt.Chart(df_theater).mark_bar().encode(
    x='Sex:N',
    y='count()',
    color='Sex:N'
).facet(
    column='Student:N'
).properties(
    title='Theater Attendance by Gender and Student Status'
)
```

```python
sex_select = alt.selection_point(
    fields=['Sex'],
    bind=alt.binding_select(
        options=[None, 'Female', 'Male', 'Other'],
        labels=['All', 'Female', 'Male', 'Other'],
        name='Gender: '
    ),
    value=None
)

chart = (
    alt.Chart(df_long)
    .mark_bar()
    .encode(
        x=alt.X('Genre:N', title='Genre', axis=alt.Axis(labelAngle=0)),
        y=alt.Y(
            'count():Q',
            title='% of respondents',
            stack='normalize',
            axis=alt.Axis(format='%')
        ),
        color=alt.Color(
            'Response:N',
            title='Viewing type',
            scale=alt.Scale(
                domain=['Watched in theaters', 'Watched on streaming'],
                range=['#185FA5', '#B5D4F4']
            )
        ),
        tooltip=[
            alt.Tooltip('Genre:N'),
            alt.Tooltip('Response:N', title='Viewing type'),
            alt.Tooltip('count():Q', title='Count'),
        ]
    )
    .add_params(sex_select)
    .transform_filter(
        alt.datum.Sex == sex_select.Sex if False else  # placeholder – use transform below
        sex_select
    )
    .properties(title='Theater vs streaming by genre', width=500, height=350)
)

chart = (
    alt.Chart(df_long)
    .mark_bar()
    .encode(
        x=alt.X('Genre:N', title='Genre', axis=alt.Axis(labelAngle=0)),
        y=alt.Y(
            'count():Q',
            title='% of respondents',
            stack='normalize',
            axis=alt.Axis(format='%')
        ),
        color=alt.Color(
            'Response:N',
            title='Viewing type',
            scale=alt.Scale(
                domain=['Watched in theaters', 'Watched on streaming'],
                range=['#185FA5', '#B5D4F4']
            )
        ),
        tooltip=[
            alt.Tooltip('Genre:N'),
            alt.Tooltip('Response:N', title='Viewing type'),
            alt.Tooltip('count():Q', title='Count'),
        ]
    )
    .add_params(sex_select)
    .transform_filter(sex_select)
    .properties(title='Theater vs streaming by genre', width=500, height=350)
)

chart.save('vis1_genre_theater_pct.html')
chart.show()
```

```python
df_vis3 = df[['# of Movies Seen in Theaters in Past 3 Months', 'Streaming Services']].copy()
df_vis3.columns = ['Theater', 'Services']

df_vis3['Num Services'] = df_vis3['Services'].apply(
    lambda x: len([s.strip() for s in str(x).split(',') if s.strip()])
)

def bin_services(n):
    if n <= 1: return '1'
    elif n == 2: return '2'
    elif n == 3: return '3'
    elif n == 4: return '4'
    elif n == 5: return '5'
    else: return '6+'

df_vis3['Services Bin'] = df_vis3['Num Services'].apply(bin_services)

theater_order = ['0', '1 to 5', '6 to 10', 'more than 10']
services_order = ['1', '2', '3', '4', '5', '6+']

df_vis3['Theater'] = pd.Categorical(df_vis3['Theater'], categories=theater_order, ordered=True)
df_vis3['Services Bin'] = pd.Categorical(df_vis3['Services Bin'], categories=services_order, ordered=True)

color_scale = alt.Scale(
    domain=theater_order,
    range=['#3C3489', '#7F77DD', '#AFA9EC', '#CECBF6']
)

select_theater = alt.selection_point(fields=['Theater'], on='click', empty=True)

# --- main stacked bar ---
bar = (
    alt.Chart(df_vis3)
    .mark_bar()
    .encode(
        y=alt.Y('Services Bin:O', title='Number of streaming services', sort=services_order),
        x=alt.X('count():Q', title='Number of respondents'),
        color=alt.Color('Theater:O', title='Movies in theaters (past 3 months)',
                        sort=theater_order, scale=color_scale),
        order=alt.Order('Theater:O', sort='ascending'),
    )
    .add_params(select_theater)
    .properties(
        title='Do more streaming subscriptions mean fewer theater visits?',
        width=500, height=280
    )
)

# opacity applied as a separate layer so it doesn't pollute the tooltip
bar = bar.encode(
    opacity=alt.condition(select_theater, alt.value(1.0), alt.value(0.25))
)

# --- detail chart ---
detail = (
    alt.Chart(df_vis3)
    .mark_bar()
    .encode(
        x=alt.X('Services Bin:O', title='Number of streaming services', sort=services_order),
        y=alt.Y('count():Q', title='Respondents'),
        color=alt.Color('Theater:O', sort=theater_order, scale=color_scale, legend=alt.Legend(title='Movies in theaters (past 3 months)')),
    )
    .transform_filter(select_theater)
    .properties(
        title=alt.TitleParams('Click a segment above to filter', fontSize=12, color='gray'),
        width=500, height=160
    )
)

combined = (bar & detail)
combined.save('vis3_streaming_vs_theater.html')
combined
```
